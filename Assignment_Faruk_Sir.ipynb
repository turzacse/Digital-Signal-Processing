```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-15, 15)


# Signal with 5-unit delay
xnn = []
delay = 5
for i in n:
    j = i - delay
    if j >= 0 and j <= 10:
        xnn.append((j * j) + j)
    elif j < 0 and j >= -10:
        xnn.append(-(j * j) + j)
    else:
        xnn.append(0)


# Signal with 5-unit advance
xnnn = []
advanced = 5
for i in n:
    j = i + advanced
    if j >= 0 and j <= 10:
        xnnn.append((j * j) + j)
    elif j < 0 and j >= -10:
        xnnn.append(-(j * j) + j)
    else:
        xnnn.append(0)


# Multiply xnnn, xnn, and the index
result = np.array(xnnn) * np.array(xnn)*4

plt.stem(n, result)
plt.title('Result of Multiplying xnnn, xnn, and Index')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
nn = np.array(n*-1);
plt.stem(nn, result)
plt.title('After Folding')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
