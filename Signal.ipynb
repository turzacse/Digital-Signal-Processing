```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-10, 11)
n
```

```python
xn = []
for i in n:
    if i==0:
        xn .append(1)
    else:
         xn .append(0)
xn
```

```python
plt.stem(n,xn)
plt.title('Unit Sample Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
xnn = []
for i in n:
    if i>=0:
        xnn .append(i)
    else:
         xnn .append(0)
xnn
```

```python
plt.stem(n,xnn)
plt.title('Unit ramp Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
x = []
for i in n:
    if i>=0:
        x .append(1)
    else:
         x .append(0)
x
```

```python
plt.stem(n,x)
plt.title('Unit Step Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
