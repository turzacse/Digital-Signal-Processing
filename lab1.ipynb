```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-20,21)
n
```

```python
xn = []
for i in n:
    if i>=0 and i<=10:
        xn .append((i*i)+i)
    elif i>=-10 and i<=0:
         xn .append(i-(i*i))
    else:
         xn .append(0)
xxn = np.array(xn)
xxn
```

```python
plt.stem(n,xxn)
plt.show()
```
