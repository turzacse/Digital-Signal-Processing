```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-20,21)
n
```

```python
xn = []
for i in n:
    if i>=0 and i<=10:
        xn .append((i*i)+i)
    elif i>=-10 and i<=0:
         xn .append(i-(i*i))
    else:
         xn .append(0)
xxn = np.array(xn)
xxn
```

```python
plt.stem(n,xn)
plt.show()
```

```python
pxn = []
for i in n:
    if(i+20-5>=0 and i+20+5<41):
     pxn.append(xn[i+20-5]*xn[i+20+5])
pxn
ppxn = np.array(pxn)
ppxn
```

```python
nn = np.arange(-15,16)
nn
```

```python
plt.stem(nn,ppxn)
```

```python
nn = np.array(nn*-1)
plt.stem(nn,ppxn)
```
