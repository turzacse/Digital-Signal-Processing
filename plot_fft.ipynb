```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 100
xt = np.zeros(t)
xt[:10] =1
plt.stem(xt)
```

```python
f = np.fft.fft(xt)
plt.stem(f)
```

```python
plt.subplot(1,2,1)
plt.plot(np.fft.fftshift(np.real(f)))
plt.title("FFT Real part")

plt.subplot(1,2,2)
plt.plot(np.fft.fftshift(np.imag(f)))
plt.title("FFFT Imaginary part")
```
